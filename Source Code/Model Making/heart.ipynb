```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
from imblearn.combine import SMOTEENN
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, classification_report, roc_curve, auc
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LSTM, Bidirectional, Dropout, Layer, Multiply, Add
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import joblib

# ------------------- Data Loading -------------------
from google.colab import drive
drive.mount('/content/drive')

path = "/content/drive/MyDrive/BG12/processed_cleveland.csv"
data = pd.read_csv(path)
print(data.head())
print(data.info())
print(data.describe())
print("Missing values per column:\n", data.isnull().sum())

# ------------------- Preprocessing -------------------
X = data.drop(columns=['num'])
y = data['num']

X = X.apply(pd.to_numeric, errors='coerce')

# KNN Imputer
imputer = KNNImputer(n_neighbors=5)
X_imputed = imputer.fit_transform(X)
joblib.dump(imputer, 'imputer.pkl')

# SMOTEENN Balancing
smote_enn = SMOTEENN(random_state=42)
X_resampled, y_resampled = smote_enn.fit_resample(X_imputed, y)
print("Balanced class distribution:\n", pd.Series(y_resampled).value_counts())

# StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)
joblib.dump(scaler, 'scaler.pkl')

# Reshape for LSTM
X_scaled = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)
print("Reshaped X for LSTM:", X_scaled.shape)

# ------------------- Residual Attention -------------------
class ResidualAttention(Layer):
    def __init__(self, units):
        super(ResidualAttention, self).__init__()
        self.dense1 = Dense(units, activation='relu')
        self.dense2 = Dense(units, activation='sigmoid')
        self.multiply = Multiply()
        self.add = Add()

    def call(self, inputs):
        x = self.dense1(inputs)
        attention = self.dense2(x)
        x = self.multiply([inputs, attention])
        x = self.add([inputs, x])
        return x

# ------------------- Model Building -------------------
def build_model(input_shape):
    inputs = Input(shape=input_shape)
    x = ResidualAttention(units=input_shape[0])(inputs)
    x = Bidirectional(LSTM(128, return_sequences=False))(x)
    x = Dropout(0.4)(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.3)(x)
    outputs = Dense(1, activation='sigmoid')(x)
    model = Model(inputs, outputs)
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# ------------------- 5-Fold Cross Validation -------------------
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
fold = 1
accuracies = []
aucs = []
best_models = []

for train_index, test_index in skf.split(X_scaled, y_resampled):
    print(f"\n--- Fold {fold} ---")
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y_resampled[train_index], y_resampled[test_index]

    model = build_model((X_train.shape[1], X_train.shape[2]))
    early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
    checkpoint = ModelCheckpoint(f'model_fold{fold}.h5', save_best_only=True, monitor='val_loss')

    history = model.fit(
        X_train, y_train,
        validation_split=0.2,
        epochs=150,
        batch_size=32,
        callbacks=[early_stop, checkpoint],
        verbose=1
    )

    # Plot training curves
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'Fold {fold} Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title(f'Fold {fold} Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

    # Evaluate predictions
    y_pred_prob = model.predict(X_test).ravel()
    y_pred_class = (y_pred_prob > 0.5).astype(int)
    y_test_binary = np.where(np.array(y_test) > 0, 1, 0)  # ensure strict binary

    acc = accuracy_score(y_test_binary, y_pred_class)
    auc_score = roc_auc_score(y_test_binary, y_pred_prob)
    print(f"Fold {fold} Accuracy: {acc:.4f}, AUC: {auc_score:.4f}")

    cm = confusion_matrix(y_test_binary, y_pred_class)
    print("Confusion Matrix:\n", cm)
    report = classification_report(y_test_binary, y_pred_class)
    print(report)

    # ROC curve
    fpr, tpr, _ = roc_curve(y_test_binary, y_pred_prob)
    plt.figure(figsize=(6,5))
    plt.plot(fpr, tpr, color='blue', label=f'ROC curve (AUC = {auc_score:.4f})')
    plt.plot([0,1], [0,1], color='red', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve Fold {fold}')
    plt.legend()
    plt.show()

    accuracies.append(acc)
    aucs.append(auc_score)
    best_models.append(f'model_fold{fold}.h5')
    fold += 1

print(f"\nAverage Accuracy: {np.mean(accuracies):.4f}")
print(f"Average AUC: {np.mean(aucs):.4f}")

# Save best models list and preprocessing objects for frontend
joblib.dump(best_models, 'best_models_list.pkl')
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(imputer, 'imputer.pkl')

print("All preprocessing objects and best models saved for frontend prediction.")
```

```python
import pandas as pd
import numpy as np
import joblib
from tensorflow.keras.models import load_model
from tensorflow.keras.layers import Layer, Dense, Multiply, Add

# ------------------- FIXED Custom Layer Definition -------------------
class ResidualAttention(Layer):
    def __init__(self, units, **kwargs):
        super(ResidualAttention, self).__init__(**kwargs)
        self.units = units
        self.dense1 = Dense(units, activation='relu')
        self.dense2 = Dense(units, activation='sigmoid')
        self.multiply = Multiply()
        self.add = Add()

    def call(self, inputs):
        x = self.dense1(inputs)
        attention = self.dense2(x)
        x = self.multiply([inputs, attention])
        x = self.add([inputs, x])
        return x

    # Needed to SAVE + LOAD correctly
    def get_config(self):
        config = super().get_config()
        config.update({
            "units": self.units
        })
        return config


# ------------------- Load Preprocessing Objects -------------------
scaler = joblib.load('scaler.pkl')
imputer = joblib.load('imputer.pkl')
best_models_list = joblib.load('best_models_list.pkl')

# Load the best model safely
model = load_model(best_models_list[0], custom_objects={'ResidualAttention': ResidualAttention})


# ------------------- Prediction Function -------------------
def predict_heart_disease(input_dict):

    feature_order = ['age','sex','cp','trestbps','chol','fbs','restecg',
                     'thalach','exang','oldpeak','slope','ca','thal']

    input_df = pd.DataFrame([[input_dict[f] for f in feature_order]], columns=feature_order)

    # Impute
    input_imputed = imputer.transform(input_df)

    # Scale
    input_scaled = scaler.transform(input_imputed)

    # Reshape for LSTM: (batch, timesteps, features)
    input_scaled = input_scaled.reshape(1, input_scaled.shape[1], 1)

    prob = model.predict(input_scaled)[0][0]
    label = "Heart Disease DETECTED" if prob > 0.75 else "NO Heart Disease"

    return label, prob


# ------------------- Input directly -------------------
print("\nEnter patient details below:")

feature_list = ['age','sex','cp','trestbps','chol','fbs','restecg',
                'thalach','exang','oldpeak','slope','ca','thal']

input_data = {}
for f in feature_list:
    value = float(input(f"Enter {f}: "))
    input_data[f] = value
2

label, probability = predict_heart_disease(input_data)

print("\n---- FINAL RESULT ----")
print("Prediction :", label)
print("Probability:", probability)
```
